## Large integer handling in numpy

Here's something annoying that happens in numpy when working with very large integers:

```python
# Plotting the sum of cubes the "dumb" way
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 100000)
cumsum_cubes = np.cumsum(n**3)

# Plot using the matrix data
plt.plot(n, cumsum_cubes)
plt.show()
```

This happens because of integer overflow. The default integer type use by numpy is int64, which is a 64-bit integer. This is a large enough type to handle numbers up to 1e18, but by 1e19, they flip signs and look like negative numbers.

To accommodate larger numbers, we can use the uint64 type, which is a 64-bit unsigned integer. This type can handle numbers up to 1e19, but they flip signs by 2e19.


```python
# Plotting the sum of cubes the "dumb" way
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 100000, dtype=np.uint64)
cumsum_cubes = np.cumsum(n**3)

# Plot using the matrix data
plt.plot(n, cumsum_cubes)
plt.show()
```

For something closer to arbitrary precision, we can use the float128 type, which is a 128-bit floating point number. There's still an upper limit, but it's much larger than the previous types. This type can handle integers up to about 2e113.

```python
# Plotting the sum of cubes the "dumb" way
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 100000000, dtype=np.float128)
cumsum_cubes = np.cumsum(n**3)

# Plot using the matrix data
plt.plot(n, cumsum_cubes)
plt.show()
```

If you need true arbitrary precision, you'll need to use a library like `sympy` instead of `numpy`.
