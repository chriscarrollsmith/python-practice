## Large integer handling in numpy

Here's something annoying that happens in numpy when working with very large integers:

```python
# Plotting the sum of cubes the "dumb" way
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 100000)
cumsum_cubes = np.cumsum(n**3, dtype=int)

# Plot using the matrix data
plt.plot(n, cumsum_cubes)
plt.show()
```
